```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
ys = [np.array([3, 4])]
xs = [np.array([-1])]
lr_y = 0.1
lr_x = 0.1
for i in range(100):
    y = ys[-1]
    x = xs[-1]
    grad_y = np.array([y[0] - x[0], y[1]])
    grad_x = np.array([x[0] - y[0]])
    ys.append(y - lr_y * grad_y)
    xs.append(x - lr_x * grad_x)
ys = np.array(ys)
xs = np.array(xs)
```

```python
plt.scatter(ys[:, 0], ys[:, 1], label="y")
plt.scatter(xs[:, 0], np.zeros_like(xs[:, 0]), c='r', label="x")
```
